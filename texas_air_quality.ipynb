```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data
df = pd.read_csv('texas_air_quality.csv')
```

```python
# Convert the 'Date' column to a datetime object
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')

# Set the 'Date' column as the index
df.set_index('Date', inplace=True)

# Create a new column with the year
df['Year'] = df.index.year
```

```python
# Create a new column with the month
df['Month'] = df.index.month

# Compute the mean air quality index for each year
aqi_by_year = df.groupby('Year')['AQI'].mean()

# Compute the mean air quality index for each month
aqi_by_month = df.groupby('Month')['AQI'].mean()
```

```python
# Plot the mean air quality index by year
plt.plot(aqi_by_year.index, aqi_by_year.values)
plt.title('Mean Air Quality Index in Texas by Year')
plt.xlabel('Year')
plt.ylabel('Air Quality Index')
plt.show()
```

```python
# Plot the mean air quality index by month
plt.plot(aqi_by_month.index, aqi_by_month.values)
plt.title('Mean Air Quality Index in Texas by Month')
plt.xlabel('Month')
plt.ylabel('Air Quality Index')
plt.show()
```
